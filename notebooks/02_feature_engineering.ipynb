```python
import pandas as pd

df = pd.read_csv("../data/atlantic_spain_cleaned.csv")

df['date'] = pd.to_datetime(df['date'])
```

```python
print("Unique Songs :", df['song_id'].nunique())
```

```python
entry_date = (
    df.groupby('song_id')['date']
      .min()
      .rename("entry_date")
)
```

```python
exit_date = (
    df.groupby('song_id')['date']
      .max()
      .rename("exit_date")
)
```

```python
lifecycle = pd.concat(
    [entry_date, exit_date],
    axis=1
)

lifecycle.head()
```

```python
lifecycle['days_on_playlist'] = (
    lifecycle['exit_date'] -
    lifecycle['entry_date']
).dt.days + 1
```

```python
best_rank = (
    df.groupby('song_id')['position']
      .min()
      .rename("best_rank")
)

lifecycle = lifecycle.join(best_rank)
```

```python
worst_rank = (
    df.groupby('song_id')['position']
      .max()
      .rename("worst_rank")
)

lifecycle = lifecycle.join(worst_rank)
```

```python
avg_rank = (
    df.groupby('song_id')['position']
      .mean()
      .rename("avg_rank")
)

lifecycle = lifecycle.join(avg_rank)
```

```python
peak_popularity = (
    df.groupby('song_id')['popularity']
      .max()
      .rename("peak_popularity")
)

lifecycle = lifecycle.join(peak_popularity)
```

```python
avg_popularity = (
    df.groupby('song_id')['popularity']
      .mean()
      .rename("avg_popularity")
)

lifecycle = lifecycle.join(avg_popularity)
```
